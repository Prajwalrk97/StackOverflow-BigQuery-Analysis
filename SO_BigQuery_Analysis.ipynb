### Query to list the 10 most popular tags from the stackoverflow dataset for the debut year, in the descending order of counts.

```python

query_string = """
    #standardSQL
    SELECT tag, COUNT(*) freq
    FROM (
      SELECT SPLIT(tags, '|') tags
      FROM `bigquery-public-data.stackoverflow.posts_questions` questions
      WHERE EXTRACT(YEAR FROM creation_date) = 2008 ),
      UNNEST(tags) tag
    GROUP BY tag
    ORDER BY freq DESC
    LIMIT 10"""

results = bigquery_client.query(query_string).result()  # Waits for job to complete.

print("Top 10 most popular tags of 2008")
for row in results:
    print("{}: {}".format(row.tag, row.freq))
```

```python
#store the above query results in data frame
Popular_Tags_2008 = (
    bigquery_client.query(query_string)
    .result()
    .to_dataframe(bqstorage_client=bqstorageclient)
)

Popular_Tags_2008
```

```python
#Some visualizations
plt.figure(figsize=(12, 9))
sns.barplot(x="tag", y="freq", data=Popular_Tags_2008, palette="coolwarm").set(
    xlabel='Tags', ylabel='Frequency', title="Popular tags of 2008");
```

### Obtaining the results of above the query for the years ranging from 2008 to 2021, into a Pandas data frame.

```python
#create a new Data frame to store the results of query as year ranges from 2008 to 2021
col_names = ['Year', 'Tag', 'Frequency']
Popular_Tags = pd.DataFrame(columns = col_names) 
list_of_tags_yearly =[]
```

```python
#iterate over years 2008 to 2021
print("Per year view of popular tags")

#subplots to store yearly plots of popular tags.
fig, axes =plt.subplots(7,2, figsize=(20,40), sharex=False);
axes = axes.flatten()

for ax, year in zip (axes, range(2008,2022)):
    #query to list the 10 most popular tags as years range from 2008 to 2021
    query_string = f"""
        #standardSQL
        SELECT Year, Tag, COUNT(*) Frequency
        FROM (
          SELECT SPLIT(tags, '|') tags, EXTRACT(YEAR FROM creation_date) Year
          FROM `bigquery-public-data.stackoverflow.posts_questions` questions
          WHERE EXTRACT(YEAR FROM creation_date) = {year} ),
          UNNEST(tags) Tag
        GROUP BY Tag, Year
        ORDER BY Frequency DESC
        LIMIT 10"""
    
    #store the results in data frame
    Popular_Tags_Of_Year = (
        bigquery_client.query(query_string)
        .result()
        .to_dataframe(bqstorage_client=bqstorageclient)
    )
    
    #Bar plot of popular tags of each year
    figure = sns.barplot(x="Tag", y="Frequency", data=Popular_Tags_Of_Year, palette="coolwarm", ax=ax);
    figure = figure.set(xlabel='Tags', ylabel='Frequency', title="Popular tags of " + str(year));
    
    #append yearly results in Popular_Tags data frame
    Popular_Tags = pd.concat([Popular_Tags, Popular_Tags_Of_Year], ignore_index=True, sort=True)
    
    #for another way of visualization - bar plot race
    list_of_tags_yearly.append(Popular_Tags_Of_Year.pivot(index='Year',columns='Tag')['Frequency'])


plt.tight_layout()  
plt.show()

#fill list_of_tags_yearly with 0 for tags which did not make to top 10 in other years
Popular_Tags_Flattened = pd.concat(list_of_tags_yearly).fillna(0)
```

### Extracting the ten highest viewcounts over the entire duration since the inception of SOF in 2008. 

```python
# Query to find top 10 posts with highest viewcounts in 2018-2021 duration
query_string = """
    #standardSQL
    SELECT
      posts.id,
      EXTRACT(YEAR FROM posts.creation_date) AS Year,
      posts.title,
      posts.view_count,
      posts.owner_user_id,
      users.display_name,
      users.location
    FROM
      `bigquery-public-data.stackoverflow.stackoverflow_posts` posts,
      `bigquery-public-data.stackoverflow.users` users
    WHERE
      posts.owner_user_id = users.id
      AND EXTRACT(YEAR FROM posts.creation_date) BETWEEN 2008 AND 2021
    ORDER BY
      view_count DESC
    LIMIT 10 """
```

### It's worth noticing that the post with the highest view count is displayed as "Unknown user". One reason could possibly be that the account has been deleted/deactivated.
<br></br>
### Also, close to 50% of the posts are from 2009 and two of the top 10 users are from India.

```python
Posts_By_View_Counts = (
    bigquery_client.query(query_string)
    .result()
    .to_dataframe(bqstorage_client=bqstorageclient)
)

Posts_By_View_Counts
```

### Extracting data year-wise between 2008 and 2021.

```python
#create a new Data frame to store the results of query as year ranges from 2008 to 2021
col_names = ['id','creation_date','title','view_count','owner_user_id','display_name','location', 'tags','Year']
Popular_Posts = pd.DataFrame(columns = col_names) 
```

```python
#iterate over years 2008 to 2021
#note: added posts.tags and Year column as well to query for extra insights

for year in range(2008,2022):
    query_string = f"""
        #standardSQL
        SELECT
          posts.id,
          EXTRACT(YEAR FROM posts.creation_date) AS Year,
          posts.title,
          posts.view_count,
          posts.owner_user_id,
          users.display_name,
          users.location,
          posts.tags   
        FROM
          `bigquery-public-data.stackoverflow.stackoverflow_posts` posts,
          `bigquery-public-data.stackoverflow.users` users
        WHERE
          posts.owner_user_id = users.id
          AND EXTRACT(YEAR FROM posts.creation_date) = {year}
        ORDER BY
          view_count DESC
        LIMIT 10"""
    
    #store the results in data frame
    Popular_Posts_Of_Year = (
        bigquery_client.query(query_string)
        .result()
        .to_dataframe(bqstorage_client=bqstorageclient)
    )
    
    #Print the yearly results
    print ("Popular posts of ", year)
    Popular_Posts_Of_Year

    #append yearly results in Popular_Posts data frame
    Popular_Posts = pd.concat([Popular_Posts, Popular_Posts_Of_Year], ignore_index=True, sort=True)
    
```

* Early years have questions with high view counts pertaining to Javascript, SQL, HTML and GIT. The later years reference questions on Java, Andriod and iOS.
